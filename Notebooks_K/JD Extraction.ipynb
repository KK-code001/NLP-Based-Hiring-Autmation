```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../Datasets/resumeJD2_pairs.csv")
if df is False:
    print("The dataset is empty.")
else:
    print("-------------------------------Dataset loaded successfully!----------------------------------")
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    print(df.columns)
    print(df.head())
print("-------------------------------Dataset Information----------------------------------")
df.info()
df.describe()
df.isnull().sum()
```

```python
print(df['job_description'][0])
patterns = {
    "job_title": r"Job Title:\s*(.*?)\s*\|\|",
    "company": r"Company:\s*(.*?)\s*\|",
    "location": r"Location:\s*(.*?)\s*\|",
    "employment_type": r"Type:\s*(.*?)\s*\|\|",
    "experience": r"Experience:\s*(.*?)\s*\|",
    "salary": r"Salary:\s*(.*?)\s*\|\|",
    "team_size": r"Team Size:\s*(.*?)\s*\|",
    "interview_rounds": r"Interview Rounds:\s*(.*?)\s*\|\|",
    "required_skills": r"REQUIRED SKILLS:\s*(.*?)\s*\|\|\s*NICE TO HAVE:",
    "nice_to_have": r"NICE TO HAVE:\s*(.*?)\s*\|\|\s*---",
    "perks": r"PERKS:\s*(.*?)\s*\|\|\s*Domain:",
    "domain": r"Domain:\s*(.*)$"
}
for field, pattern in patterns.items():
    df[field] = df["job_description"].str.extract(pattern,expand=False)
print(df.columns.tolist())
```

```python
df.drop(columns=["job_description",'resume_text','match_score','match_label'], inplace=True)
print(df.columns.tolist())
print(df.head())
df.to_csv("../DataExtracted/JD Extracted.csv",index=False)
```
