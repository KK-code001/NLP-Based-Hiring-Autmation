```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1=pd.read_csv("../Datasets/resumeJD2_pairs.csv")
df1.insert(0,"id",range(1,len(df1)+1))
print(df1.columns.tolist())
df2=pd.read_csv("../DataExtracted/JD Extracted.csv")
df2.insert(0,"id",range(1,len(df2)+1))
print(df2.columns.tolist())
df3=pd.read_csv("../DataExtracted/resume_text extracted.csv")
df3.insert(0,"id",range(1,len(df3)+1))
print(df3.columns.tolist())
```

```python
df1.drop(columns=["resume_text",'job_description'],inplace=True)
df=pd.merge(df3,df2,on="id")
print(df.columns.tolist())
dffinal=pd.merge(df,df1,on="id")
dffinal.rename(columns={'location_y':'Job_Location','location_x':'Candidate_Location'},inplace=True)
dffinal.drop(columns=["id"],inplace=True)
print(dffinal.head())
print(f"Number of rows in dffinal: {dffinal.shape[0]} and number of columns: {dffinal.shape[1]}")
print(dffinal.info())
```

```python
dffinal.to_csv("../DataExtracted/Final_Merged.csv",index=False)
```
